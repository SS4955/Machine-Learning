```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
```

```python
y_predicted
```

```python
df['cluster']=y_predicted
```

```python
df.head()
```

```python
# Assuming you have already obtained cluster labels and stored them in 'cluster' column of your DataFrame 'df'
df1 = df[df['cluster'] == 0]
df2 = df[df['cluster'] == 1]
df3 = df[df['cluster'] == 2]

plt.scatter(df1['Age'], df1['Income($)'])  # Scatter plot for cluster 0
plt.scatter(df2['Age'], df2['Income($)'])  # Scatter plot for cluster 1
plt.scatter(df3['Age'], df3['Income($)'])  # Scatter plot for cluster 2
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.title('Clusters of customers')
plt.legend(['Cluster 0', 'Cluster 1', 'Cluster 2'])
plt.show()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
```

```python
df
```

```python

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
print(y_predicted)
```

```python
df['cluster1'] = y_predicted
df.drop('cluster', axis=1, inplace=True)
```

```python
df
```

```python
df1 = df[df['cluster1'] == 0]
df2 = df[df['cluster1'] == 1]
df3 = df[df['cluster1'] == 2]

plt.scatter(df1['Age'], df1['Income($)'])  # Scatter plot for cluster 0
plt.scatter(df2['Age'], df2['Income($)'])  # Scatter plot for cluster 1
plt.scatter(df3['Age'], df3['Income($)'])  # Scatter plot for cluster 2
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.title('Clusters of customers')
plt.legend(['Cluster 0', 'Cluster 1', 'Cluster 2'])
plt.show()
```
